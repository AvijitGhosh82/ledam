```python
def ke(ent):
    return ent[0]
```

```python
documents = []
phrases = []
train_data = []
for i in range(len(os.listdir(train_doc_path))):
        doc1 = open(os.path.join(train_doc_path,'case_'+str(i)+'_statement.txt'))
        doc2 = open(os.path.join(train_catch_path,'case_'+str(i)+'_catchwords.txt'))
        try:
                doc1 = clean(doc1.read())
                doc2 = doc2.read()
                documents.append(doc1)
                phrases.append(doc2.lower().split(', '))
                entity_list = []
                print(len(doc2.lower().split(', ')))
                #for doc1 in doc1L:
                for phrase in doc2.lower().split(', '):
                        sentences = doc1.lower()
                        while(1):
                                ind = sentences.find(phrase)
                                if(ind is -1):
                                        break
                                ent_val = (ind,ind+len(phrase),'LEGAL')
                                entity_list.append(ent_val)
                                sentences = sentences[ind+len(phrase):]
                entity_list.sort(key=ke)
                print(len(entity_list))
                doc_rep = (doc1,{'entities':entity_list})
                train_data.append(doc_rep)
        except:
                print("utf-error in loading Document: {}".format(i))
```
